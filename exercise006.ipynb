```python
import random, math
import matplotlib.pyplot as plt
```

```python
def generator(a,b):
    return math.trunc(random.uniform(a,b))
```

```python
def random_ssn():
    return [generator(0,9) for i in range(9)]
```

```python
def probability_palindrome(times):
    count = 0
    for i in range(times):
        sample_ssn = random_ssn()
        reverse_ssn = list(reversed(sample_ssn))
#         print(sample_ssn)
        if sample_ssn == reverse_ssn:
            print(sample_ssn)
            count += 1
    return count/times
```

```python
probability_palindrome(1000000)
```
